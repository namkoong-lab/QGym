```python
import json
import numpy as np

envs = [
    # Reentrant
    "reentrant_2", "reentrant_3", "reentrant_4", "reentrant_5", "reentrant_6", "reentrant_7",
    "reentrant_8", "reentrant_9", "reentrant_10",
    
    # Reentrant Hyper
    "reentrant_2_hyper", "reentrant_3_hyper", "reentrant_4_hyper", "reentrant_5_hyper",
    "reentrant_6_hyper", "reentrant_7_hyper",
    
    # Re-reentrant
    "re-reentrant_2", "re-reentrant_3", "re-reentrant_4", "re-reentrant_5", "re-reentrant_6", "re-reentrant_7",
    "re-reentrant_8", "re-reentrant_9", "re-reentrant_10",
    
    # Re-reentrant Hyper
    "re-reentrant_2_hyper", "re-reentrant_3_hyper", "re-reentrant_4_hyper",
    "re-reentrant_5_hyper", "re-reentrant_6_hyper", "re-reentrant_7_hyper",

    'new-reentrant_2', 'new-reentrant_3', 'new-reentrant_4', 'new-reentrant_5', 'new-reentrant_6', 
    'new-reentrant_7', 'new-reentrant_8', 'new-reentrant_9', 'new-reentrant_10', 'new-reentrant_2_hyper', 
    'new-reentrant_3_hyper', 'new-reentrant_4_hyper', 'new-reentrant_5_hyper', 'new-reentrant_6_hyper', 
    'new-reentrant_7_hyper', 'new-re-reentrant_2', 'new-re-reentrant_3', 'new-re-reentrant_4', 
    'new-re-reentrant_5', 'new-re-reentrant_6', 'new-re-reentrant_7', 'new-re-reentrant_8', 
    'new-re-reentrant_9', 'new-re-reentrant_10', 'new-re-reentrant_2_hyper', 'new-re-reentrant_3_hyper', 
    'new-re-reentrant_4_hyper', 'new-re-reentrant_5_hyper', 'new-re-reentrant_6_hyper', 'new-re-reentrant_7_hyper',

    "hospital", "input_switch", "new_criss_cross_bh", "n_model_5x5_varying", "new_n_model"
    ]

algos = [
    'cmu', 'cmuq', 'max_pressure', 'fluid'
]


result_dict = {
    "name": [],
    'cmu': [],
    'cmuq': [],
    'fluid': [],
    'max_pressure': []
}


for env in envs:
    result_dict['name'].append(env)
    for algo in algos:
        loss_file = f"../logs/loss/{env}_{algo}/ppg/loss.json"

        try:
            with open(loss_file, 'r') as f:
                loss = json.load(f)
                result_dict[algo].append(f"${loss[0]['test_loss']:.3f} \pm {loss[0]['test_loss_std'] / np.sqrt(10):.4f}$")
        except:
            result_dict[algo].append('NA')
```

```python
import pandas as pd


df = pd.DataFrame(result_dict)
# df.columns = ['name', 'cmu', 'MW', 'MP']

df.to_csv('result_table.csv', index=False)
```
