```python
import json
import matplotlib.pyplot as plt
import pandas as pd

envs_names = ['n_model', 'criss_cross_bh', 'reentrant_2', 'reentrant_3', 'reentrant_4', 'reentrant_5', 'reentrant_6', 'reentrant_7']

fig, axs = plt.subplots(len(envs_names), 1, figsize=(10, 5 * len(envs_names)))

for i, env_name in enumerate(envs_names):
    file_path = '/user/hc3295/queue-learning/loss/' + env_name + '_ppg.json'
    with open(file_path, 'r') as file:
        data = json.load(file)
    df = pd.DataFrame(data)
    ax = axs[i]
    ax.plot(df['epoch'], df['test_loss'], label='test_loss')
    ax.fill_between(df['epoch'], df['test_loss'] - df['test_loss_std'], df['test_loss'] + df['test_loss_std'], alpha=0.3)
    ax.set_title(env_name)

plt.tight_layout()
plt.show()
```
